## Data Processing and Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.losses import MeanSquaredLogarithmicError
#Importing necessary libraries for data preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import RepeatedKFold, train_test_split
```

```python
df = pd.read_csv('ce889_dataCollection.csv', names = ['X distance to target', 'Y distance to target', 'Velocity X', 'Velocity Y'])
df.shape
```

### Normalizing

```python
scaler = MinMaxScaler()
names = df.columns
d = scaler.fit_transform(df)
normalized_df = pd.DataFrame(d, columns =names)
normalized_df.head()
```

```python
#separating the target variable from the features
targets = df[['Velocity X', 'Velocity Y']]
features = df.drop(columns=['Velocity X', 'Velocity Y'])
```

```python
print(features.shape, targets.shape)
```

```python
#Splitting the data for training. The 80:20 ratio was employed.
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.20, random_state=0)
```

## Training

```python
hidden_units1 = 160
learning_rate = 0.001
# Creating model using the Sequential in tensorflow
def build_model_using_sequential():
    model = Sequential([
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.2),
    Dense(2, kernel_initializer='normal', activation='sigmoid')
  ])
    return model
# build the model
model = build_model_using_sequential()
```

```python
# loss function
msle = MeanSquaredLogarithmicError()
rmse = RootMeanSquaredError()
model.compile(
    loss=msle, 
    optimizer=Adam(learning_rate=learning_rate), 
    metrics=[rmse]
)
# train the model
history = model.fit(
    X_train.values, 
    y_train.values, 
    epochs=50, 
    batch_size=64,
    validation_split=0.2
)
```

```python
#Plotting the model
def plot_history(history, key):
    plt.plot(history.history[key])
    plt.plot(history.history['val_'+key])
    plt.xlabel("Epochs")
    plt.ylabel(key)
    plt.legend([key, 'val_'+key])
    plt.show()
# Plot the history
plot_history(history, 'root_mean_squared_error')
```
